## 3 Selecting via Indexing, slicing, loc and iloc functions

```python
# Selecting a single column
flower["petal_length"]
```

```python
# Selecting via [], which slices the rows.
flower[0:10]
```

```python
# Selection by labels and slicing specific rows via indexing.
flower.loc[0:10,["sepal_length","sepal_width"]]
```

```python
#Showing row slicing with start and endpoints are included with specific key/column:
flower.loc[10:15,["sepal_length", "sepal_width"]]
```

```python
# We can have just one instant as well with specific row with specific key/column
flower.loc[10,["sepal_length", "sepal_width"]]
```

```python
#Selecting by position
flower.iloc[10]  
```

```python
flower.columns
```

```python
#Slicing both row and column(first we have to give rows position and then column position seperated by comma)
flower.iloc[10:15, 0:2]
```

```python
flower.columns
```

```python
# selecting specific rows and columns with iloc.
flower.iloc[[1, 2, 4], [0, 2]]
```

```python
# For slicing rows explicitly:

flower.iloc[1:3, :]

```

```python
# for slicing column explicity:

flower.iloc[:, 1:3]
```

```python
# For getting a value explicitly:
flower.iloc[2, 3]
```

# 4 Boolean indexing

```python
# To get all the flower("petal_width column") greater than 1 value
sepal_width = flower[flower["petal_width"] > 1]
sepal_width.iloc[:,3]
```

```python
# Selecting values from a iris dataset where a boolean condition is met
# we can apply this condition only if all the values in the data type are of same
# type. If the data type is mixed then we can't apply this condition.

flower[flower > 7]

# this is why it is producing error

```
