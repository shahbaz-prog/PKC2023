```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset

kashti = sns.load_dataset("titanic")
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x = "survived", hue="sex", data=kashti)
```

```python
sns.scatterplot(x= "age", y = "fare", hue="pclass", data=kashti)
```

```python
plt.scatter(x= "age", y = "fare", c="pclass", data=kashti)
```

```python
sns.barplot(x= "sex", y = "fare", hue="pclass", data=kashti)
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins= 10,figsize=(7,7), grid=False)
```

```python
sns.boxplot(data=kashti,x="embark_town",y="age",hue="class")
```

```python
g = sns.FacetGrid(kashti, hue="pclass",aspect=4)
g.map(sns.kdeplot, 'age', fill=True)
oldest = kashti['age'].max()
g.set(xlim=(0, oldest))
g.add_legend()
```

```python
sns.stripplot(data=kashti,x="embark_town",y="age",hue="sex")
```

```python
# heat map
corr = kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Heat map of Kashti")
```

```python
sns.violinplot(data=kashti,x="embark_town",y="age",hue="sex")
```
