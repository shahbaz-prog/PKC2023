```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df = sns.load_dataset("iris")
df
```

```python

df.info()
```

```python
df.isnull().sum()
```

```python
df.sepal_length.value_counts().plot(kind="bar")
```

```python
sns.countplot(x = "species", data=df)
```

```python
sns.scatterplot( x = "species", y= "sepal_length",hue = "species",data=df)
```

```python
g = plt.scatter(x= "species", y = "sepal_width", data=df)
plt.show(g)
```

```python
sns.barplot(x= "species", y = "petal_length", hue="species", data=df)
```

```python
sns.pairplot(df, hue="species")
```

```python
df.hist(bins= 15,figsize=(7,7), grid=False)
```

```python
sns.boxplot(data=df,x="species",y="petal_length",hue="species")
```

```python
sns.boxenplot(data=df,x="species",y="petal_length",hue="species")
```

```python
g = sns.FacetGrid(df, hue="species",aspect=3)
g.map(sns.kdeplot, 'petal_length', fill=True)
g.add_legend()
```

```python
sns.stripplot(data=df,x="species",y="petal_width",hue="species")
```

```python
corr = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Heat map of Iris dataset")
```

```python
sns.violinplot(data=df,x="species",y="petal_width",hue="species")
```
