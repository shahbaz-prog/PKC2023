```python
sns.heatmap(df.isnull(),yticklabels=False, cbar = False,cmap="viridis")
```

```python
df.drop(columns="deck",inplace=True)
```

```python
df.shape
```

```python
# Visualise and explore

sns.countplot(x="survived",data=df)
```

```python
sns.barplot(df,x="survived",y="age",hue="who")
```

```python
sns.countplot(x="survived",hue="who",data=df)
```

```python
#data distribution
sns.histplot(data=df,x="age")
```

```python
sns.histplot(data=df,x="age", bins=10)
```

```python
df["age"].hist(bins=30,color="darkred",alpha = "0.7")
```

```python
sns.boxplot(y="age",x = "survived",data=df)
```

```python
sns.scatterplot(x="age",y="fare",data=df, hue="pclass")
plt.ylim(0,300)
```

```python
sns.pairplot(df,hue="survived")
```

```python
#stacked barchart
df.groupby(["pclass","sex"])
```

```python
df.groupby(["pclass","sex"])["survived"].mean().unstack().plot(kind="bar",stacked=True)
```

```python
df.groupby(["pclass","who"])["survived"].mean().unstack().plot(kind="bar",stacked=True)
```

```python
#swarmplot
sns.swarmplot(data=df,x="survived",y="age",hue="who")
```
