```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
print(tf.__version__)
```

```python
# simplets sequential layer neural network


model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

creates a sequential model using the TensorFlow Keras library. The model consists of a more than one layer with more than one unit and an input shape of more than one.

```python
# compile the model
model.compile(optimizer="sgd",loss="mean_squared_error")
```

```python
#make numpy arrays for data
xs = np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype=float)
ys= np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=float)
```

```python
# training of neural network
model.fit(xs,ys,epochs=500)
```

```python
print(model.predict([10]))
```
